```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
arr=range(100)
rdd = sc.parallelize(arr)

rdd2 = rdd.filter(lambda x : x%2==0)
rdd2.collect() 

rdd3 = rdd2.map(lambda a : (a, a*a , a*a*a) )
rdd3.collect()


depremRDD= sc.textFile("hdfs://aragorn1/user/ahmet/DepremVerileri.txt")
def splitLine(line):
    arr=line.split("\t")
    return (arr[2],arr[14],arr[7])
depremRDD2 = depremRDD.map(lambda line : splitLine(line)).sortBy(lambda a : a[1])
depremRDD2.collect()                         
```
